```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('otomoto_data.csv')
df.head()
```

```python
dfg = df[['make', 'price']]
dfg = dfg.groupby('make').count().reset_index()
dfg = dfg.rename(columns={'price':'count'})
dfg
```

```python
dff = df.copy()
dff.price = [f'{price.split()[0]}{price.split()[1]}' for price in dff['price']]
dff.price = dff['price'].astype(np.int64)
dff.loc[dff['price'] <= 40000, ['make','price']]
```
